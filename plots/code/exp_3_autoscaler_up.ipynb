```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science','ieee', 'no-latex'])

df_10nodes = pd.read_csv("../files/experiment_3/metrics_embl_inference_cluster_exp3_pod_times_10nodes.csv")
df_10nodes["Initialized"] = pd.to_datetime(df_10nodes["Initialized"])
df_10nodes["Ready"] = pd.to_datetime(df_10nodes["Ready"])
valor_minimo_inicio_10nodes = df_10nodes["Initialized"].min()
df_10nodes["added_node"] = (df_10nodes["Ready"] - valor_minimo_inicio_10nodes).dt.total_seconds()
df_10nodes = df_10nodes.sort_values(by="added_node")
df_10nodes["node_count"] = range(2, len(df_10nodes) + 2)
initial_node_10nodes = pd.DataFrame({"added_node": [0], "node_count": [1]})
df_10nodes = pd.concat([initial_node_10nodes, df_10nodes], ignore_index=True).sort_values(by="added_node")

df_20nodes = pd.read_csv("../files/experiment_3/metrics_embl_inference_cluster_exp3_pod_times_20nodes.csv")
df_20nodes["Initialized"] = pd.to_datetime(df_20nodes["Initialized"])
df_20nodes["Ready"] = pd.to_datetime(df_20nodes["Ready"])
valor_minimo_inicio_20nodes = df_20nodes["Initialized"].min()
df_20nodes["added_node"] = (df_20nodes["Ready"] - valor_minimo_inicio_20nodes).dt.total_seconds()
df_20nodes = df_20nodes.sort_values(by="added_node")
df_20nodes["node_count"] = range(2, len(df_20nodes) + 2)
initial_node_20nodes = pd.DataFrame({"added_node": [0], "node_count": [1]})
df_20nodes = pd.concat([initial_node_20nodes, df_20nodes], ignore_index=True).sort_values(by="added_node")

df_30nodes = pd.read_csv("../files/experiment_3/metrics_embl_inference_cluster_exp3_pod_times_30nodes.csv")
df_30nodes["Initialized"] = pd.to_datetime(df_30nodes["Initialized"])
df_30nodes["Ready"] = pd.to_datetime(df_30nodes["Ready"])
valor_minimo_inicio_30nodes = df_30nodes["Initialized"].min()
df_30nodes["added_node"] = (df_30nodes["Ready"] - valor_minimo_inicio_30nodes).dt.total_seconds()
df_30nodes = df_30nodes.sort_values(by="added_node")
df_30nodes["node_count"] = range(2, len(df_30nodes) + 2)
initial_node_30nodes = pd.DataFrame({"added_node": [0], "node_count": [1]})
df_30nodes = pd.concat([initial_node_30nodes, df_30nodes], ignore_index=True).sort_values(by="added_node")

plt.figure()

plt.plot(df_10nodes["added_node"], df_10nodes["node_count"], drawstyle='steps-post', marker='o', markersize=2, linestyle='-', color='b', label="10")
plt.plot(df_20nodes["added_node"], df_20nodes["node_count"], drawstyle='steps-post', marker='s', markersize=2, linestyle='--', color='r', label="20")
plt.plot(df_30nodes["added_node"], df_30nodes["node_count"], drawstyle='steps-post', marker='^', markersize=2, linestyle=':', color='g', label="30")

plt.xlabel("Time (sec)")
plt.ylabel("Nodes")
plt.title("Node provisioning time upon autoscaler request")
plt.legend()
plt.legend(title='Requested nodes', bbox_to_anchor=(1.05, 0.5), loc='center left', borderaxespad=0.)

plt.grid(True)
plt.ylim(0)
plt.xlim(0)
plt.show()
```
