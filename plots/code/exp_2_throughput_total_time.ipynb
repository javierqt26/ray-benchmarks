```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scienceplots

plt.style.use(['science','ieee', 'no-latex'])

df_cluster1 = pd.read_csv("../files/experiment_2/metrics_embl_inference_cluster_exp2_2cpu_3ram.csv")
df_cluster2 = pd.read_csv("../files/experiment_2/metrics_embl_inference_cluster_exp2_3cpu_5ram.csv")
df_cluster3 = pd.read_csv("../files/experiment_2/metrics_embl_inference_cluster_exp2_6cpu_10ram.csv")

df_cluster1['Cluster'] = '2CPU 3RAM'
df_cluster2['Cluster'] = '3CPU 5RAM'
df_cluster3['Cluster'] = '6CPU 10RAM'

colors = ["green", "red", "blue"]

df_combined = pd.concat([df_cluster1, df_cluster2, df_cluster3], ignore_index=True)
```

```python
g = sns.FacetGrid(df_combined, col='Dataset', hue='Cluster', col_wrap=2, palette=colors)
g.map(sns.lineplot, 'Concurrency', 'Throughput (img/sec)', marker='o', markersize=3)
g.set_axis_labels('Concurrency (Actor)', 'Throughput (img/sec)')
g.add_legend(title='Cluster')
plt.subplots_adjust(top=0.91)
g.fig.suptitle('Throughput vs Concurrency', y=1.05, fontsize=14)

for ax in g.axes.flat:
    ax.set_title(ax.get_title())

plt.show()
```

```python
g = sns.FacetGrid(df_combined, col='Dataset', hue='Cluster', col_wrap=2, palette=colors)
g.map(sns.lineplot, 'Concurrency', 'Total time', marker='o', markersize=3)
g.set_axis_labels('Concurrency (Actor)', 'Latency (sec)')
g.add_legend(title='Cluster')
plt.subplots_adjust(top=0.91)
g.fig.suptitle('Latency vs Concurrency', y=1.05, fontsize=14)

for ax in g.axes.flat:
    ax.set_title(ax.get_title())
    
plt.show()
```
